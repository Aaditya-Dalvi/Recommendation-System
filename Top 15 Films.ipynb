```python
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
print(movies.columns)
print(credits.columns)
```

```python
df = movies.merge(credits, on = 'title')
df.head(1)
```

```python
popular_movies = df[df['vote_count'] >=5000].sort_values(by='vote_average',ascending=False)[['id','title','vote_average','vote_count','runtime','release_date']][:15]
```

```python
popular_movies.reset_index(drop=True, inplace=True)
popular_movies
```

```python
dates = popular_movies['release_date'].str.split('-')
years = []

for date in dates:
    years.append(date[0])
popular_movies['release_date']=years 
popular_movies
```
